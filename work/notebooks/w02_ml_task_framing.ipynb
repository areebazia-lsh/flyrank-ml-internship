## My Lane as an ML Task

My chosen lane is Content Refresh Prioritization.

This is a **Ranking/Scoring** problem because the goal is to rank pages based on their likelihood of needing a content refresh. Instead of simply predicting whether a page is declining, the model helps prioritize which pages should be reviewed first.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## Target or Proxy

The target is whether a page is declining, using the observed trend_direction column where "down" indicates a declining page.

This target is based on observed historical data and is used only as a decision-support signal.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## Success Metric

The success metric is Precision@50.

This measures how many of the top 50 pages selected by the model are actually declining. A higher Precision@50 means the model is better at prioritizing pages for review.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why ML Beats a Fixed Rule

A fixed rule might only use one or two features, such as impressions or content age.

An ML model can learn from multiple signals at the same time, including impressions, CTR, content age, average position, and word count. This allows it to capture more complex patterns and produce a better ranking for content refresh decisions.

The output helps content editors decide which pages should be reviewed first.
